import pandas as pd

def createDataframe(student_data: list[list[any]]) -> pd.DataFrame:
    # Define the column names to match the required schema
    column_names = ["student_id", "age"]
    
    # We pass the data and column names to the pd.DataFrame constructor
    # to return the properly formatted DataFrame
    return pd.DataFrame(student_data, columns=column_names)
