Python ä¸­çš„ zip() å‡½æ•°æ˜¯ä¸€ä¸ªéå¸¸æœ‰ç”¨ä¸”å¸¸è§çš„å†…ç½®å‡½æ•°ï¼Œå®ƒçš„ä¸»è¦ä½œç”¨æ˜¯å°†å¤šä¸ªå¯è¿­ä»£å¯¹è±¡ï¼ˆå¦‚åˆ—è¡¨ã€å…ƒç»„ã€å­—ç¬¦ä¸²ç­‰ï¼‰ä¸­å¯¹åº”ä½ç½®çš„å…ƒç´ æ‰“åŒ…æˆä¸€ä¸ªä¸ªçš„ å…ƒç»„ (tuple)ï¼Œç„¶åè¿”å›ä¸€ä¸ªå¯è¿­ä»£çš„ zip å¯¹è±¡

1. åŸºæœ¬ç”¨æ³• (æ‰“åŒ…å¯¹åº”å…ƒç´ )
å¦‚æœæ‚¨æœ‰ä¸¤ä¸ªæˆ–æ›´å¤šçš„åˆ—è¡¨ï¼Œzip() ä¼šå°†å®ƒä»¬çš„ç¬¬ä¸€ä¸ªå…ƒç´ æ‰“åŒ…åœ¨ä¸€èµ·ï¼Œç¬¬äºŒä¸ªå…ƒç´ æ‰“åŒ…åœ¨ä¸€èµ·ï¼Œä¾æ­¤ç±»æ¨ï¼š

```python
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 22]
cities = ['New York', 'London', 'Paris']

# ä½¿ç”¨ zip() æ‰“åŒ…
zipped_data = zip(names, ages, cities)

# éå† zip å¯¹è±¡ï¼Œæ¯æ¬¡è¿­ä»£å¾—åˆ°ä¸€ä¸ªå…ƒç»„
for item in zipped_data:
    print(item)

# è¾“å‡ºï¼š
# ('Alice', 25, 'New York')
# ('Bob', 30, 'London')
# ('Charlie', 22, 'Paris')
```

å°† zip å¯¹è±¡è½¬æ¢ä¸ºåˆ—è¡¨æˆ–å­—å…¸
zip() è¿”å›çš„æ˜¯ä¸€ä¸ªè¿­ä»£å™¨ (iterator)ï¼Œé€šå¸¸æ‚¨éœ€è¦å°†å…¶è½¬æ¢ä¸ºåˆ—è¡¨æˆ–å­—å…¸æ‰èƒ½çœ‹åˆ°æˆ–é‡å¤ä½¿ç”¨æ•°æ®ï¼š

```python
# è½¬æ¢ä¸ºåˆ—è¡¨ (List of Tuples)
list_of_data = list(zip(names, ages))
# list_of_data æ˜¯ [('Alice', 25), ('Bob', 30), ('Charlie', 22)]

# è½¬æ¢ä¸ºå­—å…¸ (Dictionary)
# å¸¸ç”¨åœºæ™¯ï¼šå°†ä¸¤ä¸ªåˆ—è¡¨è½¬æ¢ä¸ºé”®å€¼å¯¹
keys = ['a', 'b', 'c']
values = [1, 2, 3]
my_dict = dict(zip(keys, values))
# my_dict æ˜¯ {'a': 1, 'b': 2, 'c': 3}
```
